```python
import speech_recognition as sr
import pyttsx3
import webbrowser
import wikipedia
import pywhatkit
import datetime
import time
```

```python
engine = pyttsx3.init()
engine.setProperty('rate', 170)

def speak(text):
    print("Synthea:", text)
    engine.say(text)
    engine.runAndWait()
```

```python
# ===============================
# IMPROVED VOICE LISTENER
# ===============================

recognizer = sr.Recognizer()

def listen():
    try:
        with sr.Microphone(device_index=1) as source:   # 🔥 MIC INDEX = 1
            print("Listening...")
            recognizer.adjust_for_ambient_noise(source, duration=1.5)
            audio = recognizer.listen(
                source,
                timeout=8,
                phrase_time_limit=6
            )
            command = recognizer.recognize_google(
                audio,
                language="en-IN"
            )
            print("You said:", command)
            return command.lower()

    except sr.WaitTimeoutError:
        print("No speech detected...")
        return None
    except sr.UnknownValueError:
        print("Could not understand...")
        return None
    except Exception as e:
        print("Error:", e)
        return None
```

```python
def is_wake_word(command):
    wake_words = ["hey synthea", "synthea", "hey cynthia", "hey sindhiya"]

    for word in wake_words:
        if word in command:
            return True
    return False
```

```python
def process_command(command):

    print("Processing:", command)

    command = command.lower().strip()

    # ------------------------
    # PLAY ON YOUTUBE (FIRST)
    # ------------------------
    if "play" in command:
        song = command.replace("play", "").replace("on youtube", "").strip()

        if song == "":
            speak("Please tell me what to play.")
            return True

        speak(f"Opening YouTube for {song}")
        webbrowser.open("https://www.youtube.com/results?search_query=" + song.replace(" ", "+"))
        return True


    # ------------------------
    # WHO IS
    # ------------------------
    if "who is" in command:
        person = command.replace("who is", "").strip()

        if person == "":
            speak("Please say the full name.")
            return True

        speak(f"Searching information about {person}")
        try:
            info = wikipedia.summary(person, sentences=2)
            speak(info)
        except:
            speak("Sorry, I could not find information.")
        return True


    # ------------------------
    # GOOGLE SEARCH
    # ------------------------
    if "search" in command:
        query = command.replace("search", "").strip()
        speak("Searching on Google")
        webbrowser.open("https://www.google.com/search?q=" + query.replace(" ", "+"))
        return True


    # ------------------------
    # TIME
    # ------------------------
    if "time" in command:
        current_time = datetime.datetime.now().strftime("%I:%M %p")
        speak(f"The time is {current_time}")
        return True


    # ------------------------
    # FALLBACK WIKIPEDIA
    # ------------------------
    if len(command.split()) <= 4:
        speak(f"Searching information about {command}")
        try:
            info = wikipedia.summary(command, sentences=2)
            speak(info)
        except:
            speak("Sorry, I could not find information.")
        return True


    # ------------------------
    # STOP
    # ------------------------
    if "stop" in command:
        speak("Shutting down assistant.")
        return False


    speak("I did not understand that command.")
    return True
```

```python
print("Synthea Voice Assistant Started 🔥")
speak("Voice assistant ready. Say Hey Synthea to activate.")

assistant_active = False

while True:

    command = listen()

    if not command:
        continue

    # Wake word
    if not assistant_active:
        if is_wake_word(command):
            assistant_active = True
            speak("Hello Soham. How may I help you?")
        continue

    # Process actual command
    keep_running = process_command(command)

    if not keep_running:
        break

    assistant_active = False
```
