```python
# ============================================
# IMPORTS
# ============================================

import cv2
import requests
import datetime
import geocoder

# ============================================
# TELEGRAM CONFIG
# ============================================

BOT_TOKEN = ""
CHAT_ID = ""

# ============================================
# EMERGENCY FUNCTION (FIXED)
# ============================================

def send_telegram_alert_with_photo(frame):

    if frame is None:
        print("❌ Frame is empty")
        return

    # 📍 Location
    try:
        g = geocoder.ip('me')
        location = f"{g.city}, {g.country}" if g.city else "Location unavailable"
    except:
        location = "Location unavailable"

    # ⏰ Time
    current_time = datetime.datetime.now().strftime("%I:%M %p")

    caption = f"""
🚨 EMERGENCY ALERT 🚨

User: Soham
Location: {location}
Time: {current_time}

Live camera snapshot attached.
"""

    # 📸 Save Image
    image_path = "emergency.jpg"
    success = cv2.imwrite(image_path, frame)

    if not success:
        print("❌ Image save failed")
        return

    # 📤 Send to Telegram
    url = f"https://api.telegram.org/bot{BOT_TOKEN}/sendPhoto"

    try:
        with open(image_path, "rb") as photo:
            response = requests.post(
                url,
                data={"chat_id": CHAT_ID, "caption": caption},
                files={"photo": photo}
            )

        if response.status_code == 200:
            print("✅ Emergency Alert Sent")
        else:
            print("❌ Telegram Error:", response.text)

    except Exception as e:
        print("❌ Exception:", e)


# ============================================
# CAMERA LOOP (FIXED)
# ============================================

cap = cv2.VideoCapture(1)   # iPhone / DroidCam index

if not cap.isOpened():
    raise Exception("❌ Camera not detected")

print("🚨 Emergency System Running")
print("Press 'E' → Send Alert | 'Q' → Quit")

while True:

    ret, frame = cap.read()

    if not ret or frame is None:
        print("❌ Frame capture failed")
        continue

    cv2.imshow("Emergency Camera", frame)

    key = cv2.waitKey(1) & 0xFF

    # 🔥 SEND ALERT
    if key == ord('e'):
        print("🚨 Sending Emergency Alert...")
        send_telegram_alert_with_photo(frame)

    # ❌ EXIT
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("System Stopped.")
```
