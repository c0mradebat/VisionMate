```python
# ===============================
# IMPORT LIBRARIES
# ===============================
import cv2
import time
import pyttsx3
from deepface import DeepFace
```

```python
# ===============================
# TEXT TO SPEECH SETUP
# ===============================

engine = pyttsx3.init()
engine.setProperty('rate', 170)

def speak(text):
    print("AI:", text)
    engine.say(text)
    engine.runAndWait()
```

```python
# ===============================
# EMOTION ANALYSIS FUNCTION
# ===============================

def analyze_emotion(frame):
    try:
        result = DeepFace.analyze(
            frame,
            actions=['emotion'],
            enforce_detection=False
        )

        emotion = result[0]['dominant_emotion']
        return emotion

    except:
        return None
```

```python
# ===============================
# LIVE EMOTION DETECTION + SPEAK
# ===============================

cap = cv2.VideoCapture(1)

if not cap.isOpened():
    raise Exception("Camera not detected ❌")

print("Emotion Detection Started 🔥")

last_spoken_time = 0
SPEAK_INTERVAL = 5   # seconds gap so it doesn't spam

while True:

    ret, frame = cap.read()
    if not ret:
        break

    emotion = analyze_emotion(frame)

    if emotion:
        cv2.putText(frame,
                    f"Emotion: {emotion}",
                    (30,50),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    (0,255,0),
                    2)

        current_time = time.time()

        if current_time - last_spoken_time > SPEAK_INTERVAL:
            speak(f"Person seems {emotion}")
            last_spoken_time = current_time

    cv2.imshow("DeepFace Emotion Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("Emotion Detection Stopped.")
```
